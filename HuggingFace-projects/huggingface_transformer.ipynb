```python
from transformers import pipeline

# Explicitly use PyTorch framework
classifier = pipeline("sentiment-analysis", framework="pt")

# Analyze text
result = classifier("he was very happy")
print(result)
```

```python
from transformers import pipeline

# Create text generation pipeline (using a small model for CPU)
generator = pipeline("text-generation", model="distilgpt2", framework="pt")

prompt = "In a world where robots and humans coexist,"

story = generator(
    prompt,
    max_length=200,
    temperature=0.1,          # More creative
    top_p=0.9,
    do_sample=True,
    num_return_sequences=1
)

print(story[0]['generated_text'])
```

```python
from transformers import pipeline

# Smaller code generation model
code_gen = pipeline("text-generation", model="Salesforce/codegen-350M-mono", framework="pt")

prompt = "def calculate_fibonacci(n):"

code = code_gen(
    prompt,
    max_length=100,
    temperature=0.2,          # Very focused for code
    do_sample=True
)

print(code[0]['generated_text'])
```
